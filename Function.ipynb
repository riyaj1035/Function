7. What are the advantages of using generators over regular functions?
 -  Advantages
 -  Memory Efficiency
     -  Generators produce one value at a time instead of storing all values in memory.
     -  Suitable for large datasets or infinite sequences.
 -  Lazy Evaluation
      -  Values are generated only when needed.
      -  Improves performance by avoiding unnecessary computation.
 -  Better Performance
      -  Faster execution for large iterations since data is not preloaded.
      -  Reduces overhead compared to building large lists.

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 -  map()
     -  Applies a function to every element of an iterable.
     -  Returns a new iterable (map object) with transformed values.
 -  reduce()
     -  Repeatedly applies a function to the elements.
     -  Reduces the iterable to a single value (like summing or multiplying elements).
     -  Available in functools module.
 -  filter()
     -  Applies a function that returns True/False.
     -  Keeps only the elements for which the function returns True.
     -  Returns a filter object (with elements that satisfy the condition).

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
# Example usage
my_list = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(my_list))
```

```python
#2.Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(text):
    return text[::-1]
# Example usage
input_string = "Hello"
print(reverse_string(input_string))
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_list(numbers):
    result = []
    for num in numbers:
        result.append(num ** 2)
    return result
# Example usage
my_list = [1, 2, 3, 4, 5]
print(square_list(my_list))
```

```python
#4.Write a Python function that checks if a given number is prime or not from 1 to 200.
def check_prime():
    num = int(input("Enter a number between 1 and 200: "))
    if num < 1 or num > 200:
        print("Number out of range! Please enter a number between 1 and 200.")
        return
    if num < 2:
        print(f"{num} is not a prime number.")
        return
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            return
    print(f"{num} is a prime number.")
check_prime()

```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_val = self.a + self.b
            self.a, self.b = self.b, next_val
            self.count += 1
            return next_val
fib = Fibonacci(10)
for num in fib:
    print(num)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
# Example usage:
for value in power_of_two(10):
    print(value)
```

```python
#7.Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(example.txt):
    try:
        with open(example.txt, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print("The specified file does not exist.")
for line in read_file_lines("example.txt"):
    print(line)
```

```python
#8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
# Sample list of tuples
data = [(1, 5), (3, 2), (4, 8), (2, 4)]

# Sorting using lambda based on second element
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
#9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# List of temperatures in Celsius
celsius = [0, 10, 20, 30, 40]
# Using map with lambda to convert to Fahrenheit
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
#10.Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Given string
text = "Hello Riyaj"
# Using filter to remove vowels
vowels = "aeiouAEIOU"
result = "".join(filter(lambda ch: ch not in vowels, text))
print("Original String:", text)
print("String without vowels:", result)
```

```python
#11(a).Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity.
#The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]
result = []
for order in orders:
    order_number = order[0]
    quantity = order[2]
    price_per_item = order[3]
    total = quantity * price_per_item
    # Add 10€ handling fee if total is less than 100€
    if total < 100:
        total += 10
    result.append((order_number, round(total, 2)))
print(result)

```

```python
#11(b).Write a Python program using lambda and map.
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]
# Using map and lambda function
result = list(map(lambda order:
                  (order[0], round(order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0), 2)),
                  orders))
print(result)
```
